```python
import torch
from torchvision import transforms
from PIL import Image
import numpy as np
import urllib.request
```

```python
urllib.request.urlretrieve(
  'https://cdn.theatlantic.com/thumbor/wCH56kkEzNtQGx51xEYg4V3lKtk=/0x57:1088x669/976x549/media/img/mt/2015/07/minions1/original.jpg',
   "gfg.png")
  
img = Image.open("gfg.png")
img
```

```python
convert_tensor = transforms.ToTensor()
img=convert_tensor(img)
img
```

```python
img = img.unsqueeze(0)
conv = torch.nn.Conv2d(3, 3, 4)
img = conv(img)
img = img.squeeze(0)
img = transforms.ToPILImage()(img)
img
```
